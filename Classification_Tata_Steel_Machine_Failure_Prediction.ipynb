# **Project Summary -**

This project focuses on predicting machine failures in a steel manufacturing environment using machine learning techniques. In large-scale industries like steel production, machine failures can lead to significant financial losses, production delays, and operational inefficiencies. Therefore, predicting failures in advance allows companies to take preventive measures and maintain smooth operations.
The dataset used in this project consists of operational parameters such as air temperature, process temperature, rotational speed, torque, and tool wear. Additionally, it includes different failure types and a target variable called "Machine Failure," which indicates whether a machine has failed.
The project begins with understanding the dataset and performing exploratory data analysis (EDA). Various visualization techniques such as count plots, histograms, box plots, and heatmaps are used to identify patterns, trends, and correlations between variables. The dataset is checked for missing values, duplicates, and outliers to ensure data quality.
Data preprocessing includes encoding categorical variables like product type and scaling numerical features using Min-Max scaling. Special care is taken to avoid data leakage by fitting transformations only on the training dataset and applying them to the test dataset.
Since the problem is a classification task, multiple machine learning models are used, including Logistic Regression and Random Forest. Among these, Random Forest performs better due to its ability to capture complex relationships in the data.
The trained model is then used to make predictions on the separate test dataset, simulating real-world deployment where unseen data is provided. The predictions are saved in a CSV file for submission.
Feature importance analysis reveals that parameters like torque, tool wear, and rotational speed are key contributors to machine failure. These insights can help businesses optimize machine usage and maintenance schedules.
Overall, this project demonstrates how machine learning can be effectively applied in industrial settings to improve efficiency, reduce downtime, and support data-driven decision-making. The solution is scalable and can be integrated into real-time monitoring systems for continuous performance

##Model Building

````python
# Train model
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X, y)
````

We trained a Random Forest model to predict machine failure.

##Prediction on Test Data

Predictions are made on unseen test data to simulate real-world performance.

##Model Evaluation

````python
from sklearn.metrics import classification_report

y_pred = model.predict(X)
print(classification_report(y, y_pred))
````

Insight:
The model shows 100% accuracy, precision, recall, and F1-score.

Important Observation:
This perfect performance indicates data leakage, as failure type features (TWF, HDF, PWF, OSF, RNF) are directly related to the target variable.

Conclusion:
The model is overfitting and not realistic. These features must be removed for proper evaluation.

##Drop leakage + useless columns

````python
drop_cols = ['id', 'Product ID', 'TWF', 'HDF', 'PWF', 'OSF', 'RNF']

train_df = train_df.drop(columns=drop_cols, errors='ignore')
test_df = test_df.drop(columns=drop_cols, errors='ignore')
````

##Encode categorical

````python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

train_df['Type'] = le.fit_transform(train_df['Type'])
test_df['Type'] = le.transform(test_df['Type'])
````

##Split features & target

````python
X = train_df.drop('Machine failure', axis=1)
y = train_df['Machine failure']
````

##Scaling

````python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_scaled = scaler.fit_transform(X)
test_scaled = scaler.transform(test_df)
````

##Model Building

````python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=42)
model.fit(X_scaled, y)
````

##Model Evaluation(Better Way)

````python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

# Proper evaluation
X_train, X_val, y_train, y_val = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

model.fit(X_train, y_train)

y_pred = model.predict(X_val)

print(classification_report(y_val, y_pred))
````

Insight:
The model achieved high overall accuracy (~99%), indicating strong performance in predicting non-failure cases.

Important Observation:

Class 0 (No Failure) → Very high precision and recall
Class 1 (Failure) → Lower recall (~33%)

##Prediction on Test Data
