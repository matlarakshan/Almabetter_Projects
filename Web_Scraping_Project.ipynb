```python
# Importing Liberaries
import requests
from bs4 import BeautifulSoup
import csv
```

```python
def scrape_data_science_jobs():
    base_url = "https://www.timesjobs.com/candidate/job-search.html"
    params = {
        "searchType": "personalizedSearch",
        "from": "submit",
        "txtKeywords": "data science",
        "txtLocation": "",
    }

    # Send a GET request to the TimesJobs website
    response = requests.get(base_url, params=params)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.content, "html.parser")

        # Extract data science job information
        job_listings = soup.find_all("li", class_="clearfix job-bx wht-shd-bx")

        # Create a CSV file and write header
        with open("data_science_jobs.csv", "w", newline="", encoding="utf-8") as csvfile:
            fieldnames = ["Job Title", "Company Name", "Experience", "Location"]
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

            writer.writeheader()

            # Write job data to CSV
            for job in job_listings:
                job_title = job.find("h2").text.strip()
                company_name = job.find("h3", class_="joblist-comp-name").text.strip()
                job_experience = job.find("ul", class_="top-jd-dtl clearfix").find("li").text.strip()
                job_location = job.find("ul", class_="top-jd-dtl clearfix").find_all("li")[1].text.strip()

                # Write row to CSV
                writer.writerow({
                    "Job Title": job_title,
                    "Company Name": company_name,
                    "Experience": job_experience,
                    "Location": job_location
                })

                print(f"Job Title: {job_title}")
                print(f"Company Name: {company_name}")
                print(f"Experience: {job_experience}")
                print(f"Location: {job_location}")
                print("="*50)

        print("Data has been successfully written to data_science_jobs.csv.")
    else:
        print(f"Failed to retrieve page. Status code: {response.status_code}")
```

```python
# Run the scraper for data science jobs and save to CSV
scrape_data_science_jobs()
```

```python
# Import Libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load Dataset
df = pd.read_csv("/content/data_science_jobs.csv")
```

```python
# Standardize text data (convert to lowercase)
df["Job Title"] = df["Job Title"].str.lower()
df["Company Name"] = df["Company Name"].str.lower()
```

```python
# Remove "location_on\n" prefix from "Location" column
df["Location"] = df["Location"].str.replace("location_on\n", "")
```

```python
# Extract numerical information from "Experience" column
df["Experience"] = df["Experience"].str.extract('(\d+)').astype(float)
```

```python
# Creating new column called salary
df["Salary"] = "N/A"
```

```python
# Extract salary information from the "Location" column
salary_extraction = df["Location"].str.extract(r'‚ÇπRs (\d+\.\d+) - (\d+\.\d+) Lacs p.a.')
```

```python
# If both lower and upper salary limits are present, use the lower limit, otherwise, use the original text
df["Salary"] = salary_extraction[0].combine_first(salary_extraction[1])
```

```python
# Remove the extracted salary information from the "Location" column
df["Location"] = df["Location"].str.replace(r'‚ÇπRs (\d+\.\d+) - (\d+\.\d+) Lacs p.a.', '').str.strip()
df["Location"].replace("", pd.NA, inplace=True)
```

```python
# Convert the "Salary" column to numeric
df["Salary"] = pd.to_numeric(df["Salary"], errors="coerce")
```

```python
# Fill null values in "Salary" column with mean and Drop rows with null values in "Location" column
df["Salary"].fillna(df["Salary"].mean(), inplace=True)
df.dropna(subset=["Location"], inplace=True)
```
