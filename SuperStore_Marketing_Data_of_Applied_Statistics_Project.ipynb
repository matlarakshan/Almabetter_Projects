##Task 2 - Descriptive Statistics üìä

- **Calculate measures of central tendency (mean, median, mode) and measures of dispersion (variance, standard deviation) for key variables. Identify and handle outliers if necessary.**


**Deliverables**:

- **Descriptive statistics that reveal the central tendencies, variations, and potential outliers in the dataset.**:

  

```python
# To show thw description of the data in the DataFrame
data.describe()
```

```python
# Calculate Mean, Median, and Mode
mean_age = data['Year_Birth'].mean()
print("mean_age is", mean_age)
median_age = data['Year_Birth'].median()
print("median_age is", median_age)
mode_age = data['Year_Birth'].mode().values[0]
print("mode_age is", mode_age)
```

```python
# Calculate Variance and Standard Deviation
variance_income = data['Income'].var()
print("variance_income is", variance_income)
std_deviation_income = data['Income'].std()
print("std_deviation_income is", std_deviation_income)
```

```python
# Boxplot to visulize the outliers
plt.figure(figsize=(10, 6))
data.boxplot(column =['Income'], grid = False)
plt.title('Box Plot of Income')
plt.show()
```

```python
# Calculate the IQR (Interquartile Range)
Q1 = data['Income'].quantile(0.25)
Q3 = data['Income'].quantile(0.75)
IQR = Q3 - Q1
print('IQR value is', IQR)
```

```python
# Define upper and lower bounds to identify outliers
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)
print('Lower fence is', lower_bound)
print('Upper fence is', upper_bound)
```

```python
# Identify and potentially remove outliers
outliers = data[(data['Income'] < lower_bound) | (data['Income'] > upper_bound)]
```

```python
# Remove outliers from the dataset
data = data[(data['Income'] >= lower_bound) & (data['Income'] <= upper_bound)]
```

```python
# Boxplot after removies outliers
data.boxplot(column =['Income'], grid = False)
plt.title('Outliers removed in Income')
```

```python
# Visualize the variable's distribution
plt.figure(figsize=(8, 6))
sns.histplot(data['NumWebPurchases'], kde=True)
plt.title('Distribution of NumWebPurchases')
plt.xlabel('NumWebPurchases')
plt.ylabel('Frequency')
plt.show()
```

```python
from scipy.stats import norm, binom
```

```python
# Fit the data to a Normal distribution
mu, std = norm.fit(data['NumWebPurchases'])
plt.figure(figsize=(8, 6))
sns.histplot(data['NumWebPurchases'], kde=True, label=f'Fit Results:\n$\mu$ = {mu:.2f}, $\sigma$ = {std:.2f}')
plt.title('Fit to Normal Distribution NumWebPurchases')
plt.xlabel('NumWebPurchases')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Fit the data to a Binomial distribution
n = max(data['NumWebPurchases'])  # Maximum number of trials
p = data['NumWebPurchases'].mean() / n  # Probability of success

# Calculate probabilities and expected values based on the Binomial distribution
probs = [binom.pmf(k, n, p) for k in range(n + 1)]
expected_value = n * p

print("Binomial Distribution Probabilities:")
for k, prob in enumerate(probs):
    print(f'P(X = {k}) = {prob:.4f}')

print(f'Expected Value (mean) = {expected_value:.2f}')
```

- **Explore relationships between customer characteristics and spending habits. Segment customers based on their behaviors and characteristics.**

```python
# Analyze Relationships: using heat map
plt.figure(figsize=(13,10))
sns.heatmap(data.corr(),annot=True)
```

```python
# Use Pairplot for visualizing relationships between numerical variables
numerical_columns = ['Year_Birth', 'Income', 'MntFishProducts', 'MntMeatProducts', 'MntFruits']
sns.pairplot(data[numerical_columns])
plt.title("Pairplot of Numerical Variables")
plt.show()
```

```python
# Select features for segmentation (e.g., spending on different product categories)
features = ['MntFishProducts', 'MntMeatProducts', 'MntFruits']
```

```python
from sklearn.preprocessing import StandardScaler
# Standardize the data

scaler = StandardScaler()
data[features] = scaler.fit_transform(data[features])
```

```python
from sklearn.cluster import KMeans
# Use K-Means clustering for customer segmentation

n_clusters = 3  # Adjust the number of clusters as needed
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
data['segment'] = kmeans.fit_predict(data[features])
```

```python
from sklearn.decomposition import PCA
# Reduce dimensionality for visualization

pca = PCA(n_components=2)
data['pca1'], data['pca2'] = pca.fit_transform(data[features]).T
```

```python
# Visualize customer segmentation in 2D space
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='pca1', y='pca2', hue='segment', palette='viridis')
plt.title("Customer Segmentation based on Spending Behavior")
plt.show()
```
